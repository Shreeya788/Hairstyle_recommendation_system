# Data Quality and Reliability

The dataset contains images of 234 celebrities. Of these, 33 celebrities had a unanimous classification from 3 or more sites (65 from 2 or more). 49 other celebrities had one or two conflicting classifications but had strong consensus towards a face shape with which I could use to classify.  As a result, 74 celebrities were utilized in the analysis.  There were ~160 celebrities I did not include as they either had too few citations or several conflicts. More labeled data would be ideal and would be an area of future enhancement. 

```python
celeb_count = [33,22,49,130]
labels = ['3+ unanimous','2 unaimous','Consensus but 1-2 conflicts','not enough info']
patches = plt.pie(celeb_count, labels = labels,autopct='%1.f%%')
plt.axis('equal')
plt.title('Break out of Celebrities by level of consensus on shape classification')
plt.tight_layout()
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter
fig, ax = plt.subplots()
formatter = FuncFormatter(lambda y, pos: "%d%%" % (y))
ax.yaxis.set_major_formatter(formatter)


percent_unanim = [33,44,52,69,74]
shapes = ['heart','long','oval','round','square']
y_pos = np.arange(len(shapes))

plt.bar(y_pos, percent_unanim, align='center', alpha=0.5)
plt.xticks(y_pos, shapes)
plt.title('% of celebrities with unanimous classification by shape')
 
plt.show()
```
